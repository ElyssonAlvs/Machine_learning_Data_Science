```python
# data the rating of movies
uri = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula1.2/ratings.csv"
# read of adress of datas with pandas
notas = pd.read_csv(uri)
# reading the first rows of datas
notas.head()
```

```python
# modifying columns names for portuguese
notas.columns = ['usuarioId', 'filmesId', 'nota', 'momento']
# reading the first rows of datas
notas.head()
```

```python
# First columns of movie notes
notas['nota'].head() 
```

```python
# unique notes presents in dataframe, metric of notes
notas['nota'].unique()
```

```python
# finding of mean of notes presents in dataFrame
notas['nota'].mean()
```

```python
# minimun note
min_notes = notas['nota'].min()
print('Minimun notes at the dataFrame "notas" : {} '.format(min_notes))
# maximun note
max_notes = notas['nota'].max()
print('Maximum notes at the dataFrame "notas" : {} '.format(max_notes))
```
